```python
def clean_data(filepath):
    data = pd.read_csv(filepath)
    
    #converting the disposable income to a standard format
    mapping = {
    'Less than 1000': 800,
    '1000 - 2000': 1600,
    '2000 - 3000': 2500,
    'More than 3000': 4000
}

    data['income'] = data['Disposable Income/Allowance'].map(mapping)
        
    #create our target column using Do you have any Life or Non-Life Insurance Policy column
    target_mapping = {
    'Yes': 1,
    'No': 0}    

    data['target'] = data['Do you have any Life or Non-Life Insurance Policy'].map(target_mapping)
    
    data['target'] = data['target'].fillna(0)
    
    data['target'] = data['target'].astype(int)
    
    # remove unimportant columns from the dataset
    leaky_cols = ['Timestamp','Username','Gender','Disposable Income/Allowance','Do you have any Life or Non-Life Insurance Policy']
    
    data = data.drop(columns = leaky_cols)
    
    return data
```

```python
def clean_data2(filepath):
    data = pd.read_csv(filepath)
    
    #converting the disposable income to a standard format
    mapping = {
    'Less than 1000': 800,
    '1000 - 2000': 1600,
    '2000 - 3000': 2500,
    'More than 3000': 4000
    }
    
    data['income'] = data['Disposable Income/Allowance'].map(mapping)
    
        
    #create our target column using Do you have any Life or Non-Life Insurance Policy column
    target_mapping = {
    'Yes': 1,
    'No': 0} 
  
    educational_mapping = {
        'Secondary':0,
        'Tertiary':1}
    
    occupational_mapping = {
        'Student':0,
        'Unemployed':1,
        'Employed':2
    }
    

    same_mapping = {
        'No':0,
        'Yes':1
    }
    
    experience_mapping = {
        'No':0,
        'Maybe':0,
        'Yes':1
    }
    
    past_experience = {
        "Bad":0,
        "Neutral":1,
        "Good":2
    }
    
    data['educ_bag'] = data['Educational Background'].map(educational_mapping)
    
    data['educ_bag'] = data['educ_bag'].astype(int)
    
    data['occ_status'] = data['Occupational Status'].map(occupational_mapping)
    
    data['state_req'] = data['Statutory Requirements'].map(same_mapping)
    data['pers_ben'] = data['Personal Benefits'].map(same_mapping)
    data['tech_ad'] = data['Technological Advancement'].map(same_mapping)
    data['rel_aff'] = data['Religious Beliefs / Affiliations'].map(same_mapping)
    data['edu_level'] = data['Educational Level'].map(same_mapping)
    data['decision_influence'] = data['Did this experience influence your perception and/or decision to buy insurance policies.'].map(experience_mapping)
    
    data['experi_impli'] = data['What is your past or current experience with insurance companies.'].map(past_experience)
    
    data['target'] = data['Do you have any Life or Non-Life Insurance Policy'].map(target_mapping)
    
    data['target'] = data['target'].fillna(0)
    
    data['target'] = data['target'].astype(int)
    
    # remove unimportant columns from the dataset
    leaky_cols = ['Disposable Income/Allowance','Do you have any Life or Non-Life Insurance Policy','Educational Background','Occupational Status','Statutory Requirements','Personal Benefits','Technological Advancement','Religious Beliefs / Affiliations','Educational Level','Did this experience influence your perception and/or decision to buy insurance policies.','What is your past or current experience with insurance companies.']
    
    data = data.drop(columns = leaky_cols)
    
    data = data.dropna()
    
    return data
```
